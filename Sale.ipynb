```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import warnings
import os
path = "D:/Python/Tu-luyen/"
df=pd.read_csv(path + 'sales2019_1.csv')
df.head()
```

```python
os.listdir(path)
```

```python
frames = []
for file in os.listdir(path):
    if file.endswith('.csv'):
        filepath = path + file
        df1=pd.read_csv(filepath)
        frames.append(df1)
result = pd.concat(frames)
df=result
df
```

```python
df['Months'] = df['Order Date'].str[0:2]
df
```

```python
print(set(df['Months']))
```

```python
df.dropna()
```

```python
df=df[df['Months'] != 'Or']
df.head()
```

```python
df['Quantity Ordered']=pd.to_numeric(df['Quantity Ordered'],downcast='integer')
df['Price Each']=pd.to_numeric(df['Price Each'],downcast= 'float')
df['Sales']=df['Quantity Ordered']*df['Price Each']
df
```

```python
moving_column = df.pop('Sales')
df.insert(4,'Sales',moving_column)
df.head()
```

```python
df.groupby('Months').sum()['Sales']
```

```python
sale_values=df.groupby('Months').sum()['Sales'] 
sale_values.max()
for month,sales in sale_values.items():
    if sales == sale_values.max():
        print(month,sales)

```

```python
month = range(1,13)
plt.bar(month,sale_values)
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Sales Report')
plt.show
```

```python
df['City'] =df['Purchase Address'].str.split(',', expand=True)[1]
df.head()
```

```python
df.groupby(by = 'City').sum()['Sales']
```

```python
sale_city=df.groupby('City').sum()['Sales'] 
sale_city.max()
```

```python
cities = [city for city,sale_city in sale_city.items()]
print(cities)
```

```python
plt.bar(x=cities,height=sale_city)
plt.xticks(cities,rotation = 90,size = 8)
plt.xlabel('City')
plt.ylabel('Sales')
plt.title('Sales by City')
plt.show()
```

```python
print(df['Order Date'].dtypes)
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
```

```python
df['Hour']=df['Order Date'].dt.hour
df
```

```python
sale_hour = df.groupby('Hour').sum()['Sales']
hours =  [hour for hour,sale_hour in sale_hour.items()]
plt.plot(hours,sale_hour)
plt.grid()
plt.xlabel('Hour')
plt.ylabel('Sale_Hour')
plt.show()
```

```python
sale_hour = df.groupby('Hour').count()['Sales']
hours =  [hour for hour,sale_hour in sale_hour.items()]
plt.plot(hours,sale_hour)
plt.grid()
plt.xlabel('Hour')
plt.ylabel('Sale_Hour')
plt.show()
```

```python
df_dup = df[df['Order ID'].duplicated(keep=False)]
```

```python
group_product = lambda product:', '.join(product)
```

```python
df_dup['All Products']=df_dup.groupby(by='Order ID')['Product'].transform(group_product)
df_dup.head()
```

```python
quan_product = df.groupby('Product').sum()['Quantity Ordered']
```

```python
prices = df.groupby('Product').mean()['Price Each']
```

```python
product = [product for product,quan_product in quan_product.items()]
plt.bar(product,quan_product)
plt.xticks(product,rotation = 90,size = 8)
plt.xlabel('Product')
plt.ylabel('Quantity_Product')
plt.show()
```

```python
x=product
y1=quan_product
y2=prices

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(x,y1,color='g')
ax2.plot(x,y2,'b-')

ax1.set_xticklabels(product,rotation=90,size = 8)
ax1.set_xlabel('Product')
ax2.set_ylabel('Quantity Products', color='g')
ax2.set_ylabel('Prices', color='b')

plt.show()
```
